```python
import math

print(math.ceil(2.3))

#ФУНКЦИЯ ПОДСЁТА ЕЖЕМЕСЯЧНОГО ДОХОДА КОМПАНИИ С ПОЛЬЗОВАТЕЛЯ
def revenue(row):
    minutes_used = row['duration']
    messages_used = row['messages_used']
    mb_used = row['mb_used']
    minutes_included = row['minutes_included']
    messages_included = row['messages_included']
    mb_per_month_included = row['mb_per_month_included']
    rub_per_minute = row['rub_per_minute']
    rub_per_message = row['rub_per_message']
    rub_per_gb = row['rub_per_gb']
    fee = row['rub_monthly_fee']
    charge = fee
    minute = minutes_used - minutes_included
    mes = messages_used - messages_included
    mb = mb_used - mb_per_month_included

    if minute > 0:
        minute = math.ceil(minute) * rub_per_minute
    else:
        minute = 0
    if mes > 0:
        mes = mes * rub_per_message
    else:
        mes = 0
    if mb > 0:
        mb = math.ceil(mb/1024) * rub_per_gb
    else:
        mb = 0
    charge += (minute + mes + mb)
    return charge
 
data_subcategory_5['revenue'] = data_subcategory_5.apply(revenue, axis=1) #добавление столбца 'revenue' 
display(data_subcategory_5)    
data_subcategory_5.info()
```

Для проверки нулевой гипотезы используется метод scipy.stats.ttest_ind (). В данном случае используется именно этот критерий, так как сравниваем средние двух совокупностей.

По итогам исследования отвергаем нулевую гипозу, <b>т.е. принимаем гипотезу H1 о том, что средняя выручка пользователей тарифов различается</b>.

Для проверки нулевой гипотезы используется метод scipy.stats.ttest_1samp (array, popmean). В данном случае используется именно этот критерий, так как сравниваем среднюю выручку по Москве(одно значение) с выручкой пользователей из других регионов(целым столбцом значений).
